```python
# có 5 trường hợp 1...5 pods
for i in range(1,6):
    file_name = "../data/25_7_2023/wifi/data/curl/mec/x86_pod_"+str(i)+"_rep_1_25_7_2023_10h43.csv"
    # NONE
    time_namelookup_none = []
    time_connect_none = []
    time_pretransfer_none = []
    time_starttransfer_none = []
    time_total_none = []
    # 360P
    time_namelookup_360P = []
    time_connect_360P = []
    time_pretransfer_360P = []
    time_starttransfer_360P = []
    time_total_360P = []
    # 480P
    time_namelookup_480P = []
    time_connect_480P = []
    time_pretransfer_480P = []
    time_starttransfer_480P = []
    time_total_480P = []
    # HD
    time_namelookup_HD = []
    time_connect_HD = []
    time_pretransfer_HD = []
    time_starttransfer_HD = []
    time_total_HD = []
    # 2K
    time_namelookup_2K = []
    time_connect_2K = []
    time_pretransfer_2K = []
    time_starttransfer_2K = []
    time_total_2K = []
    # 4K
    time_namelookup_4K = []
    time_connect_4K = []
    time_pretransfer_4K = []
    time_starttransfer_4K = []
    time_total_4K = []

    with open(file_name, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in spamreader:
            if row[8] == "NONE":
                time_namelookup_none.append(float(row[0]))
                time_connect_none.append(float(row[1]))
                time_pretransfer_none.append(float(row[3]))
                time_starttransfer_none.append(float(row[5]))
                time_total_none.append(float(row[6]))
            elif row[8] == "360P":
                time_namelookup_360P.append(float(row[0]))
                time_connect_360P.append(float(row[1]))
                time_pretransfer_360P.append(float(row[3]))
                time_starttransfer_360P.append(float(row[5]))
                time_total_360P.append(float(row[6]))
            elif row[8] == "480P":
                time_namelookup_480P.append(float(row[0]))
                time_connect_480P.append(float(row[1]))
                time_pretransfer_480P.append(float(row[3]))
                time_starttransfer_480P.append(float(row[5]))
                time_total_480P.append(float(row[6]))
            elif row[8] == "HD":
                time_namelookup_HD.append(float(row[0]))
                time_connect_HD.append(float(row[1]))
                time_pretransfer_HD.append(float(row[3]))
                time_starttransfer_HD.append(float(row[5]))
                time_total_HD.append(float(row[6]))
            elif row[8] == "2K":
                time_namelookup_2K.append(float(row[0]))
                time_connect_2K.append(float(row[1]))
                time_pretransfer_2K.append(float(row[3]))
                time_starttransfer_2K.append(float(row[5]))
                time_total_2K.append(float(row[6]))
            elif row[8] == "4K":
                time_namelookup_4K.append(float(row[0]))
                time_connect_4K.append(float(row[1]))
                time_pretransfer_4K.append(float(row[3]))
                time_starttransfer_4K.append(float(row[5]))
                time_total_4K.append(float(row[6]))

        # Mean
        mean_time_namelookup_none.append(np.mean(time_namelookup_none))
        mean_time_connect_none.append(np.mean(time_connect_none))
        mean_time_pretransfer_none.append(np.mean(time_pretransfer_none))
        mean_time_starttransfer_none.append(np.mean(time_starttransfer_none))
        mean_time_total_none.append(np.mean(time_total_none))

        mean_time_namelookup_360P.append(np.mean(time_namelookup_360P))
        mean_time_connect_360P.append(np.mean(time_connect_360P))
        mean_time_pretransfer_360P.append(np.mean(time_pretransfer_360P))
        mean_time_starttransfer_360P.append(np.mean(time_starttransfer_360P))
        mean_time_total_360P.append(np.mean(time_total_360P))

        mean_time_namelookup_480P.append(np.mean(time_namelookup_480P))
        mean_time_connect_480P.append(np.mean(time_connect_480P))
        mean_time_pretransfer_480P.append(np.mean(time_pretransfer_480P))
        mean_time_starttransfer_480P.append(np.mean(time_starttransfer_480P))
        mean_time_total_480P.append(np.mean(time_total_480P))

        mean_time_namelookup_HD.append(np.mean(time_namelookup_HD))
        mean_time_connect_HD.append(np.mean(time_connect_HD))
        mean_time_pretransfer_HD.append(np.mean(time_pretransfer_HD))
        mean_time_starttransfer_HD.append(np.mean(time_starttransfer_HD))
        mean_time_total_HD.append(np.mean(time_total_HD))

        mean_time_namelookup_2K.append(np.mean(time_namelookup_2K))
        mean_time_connect_2K.append(np.mean(time_connect_2K))
        mean_time_pretransfer_2K.append(np.mean(time_pretransfer_2K))
        mean_time_starttransfer_2K.append(np.mean(time_starttransfer_2K))
        mean_time_total_2K.append(np.mean(time_total_480P))
        
        mean_time_namelookup_4K.append(np.mean(time_namelookup_4K))
        mean_time_connect_4K.append(np.mean(time_connect_4K))
        mean_time_pretransfer_4K.append(np.mean(time_pretransfer_4K))
        mean_time_starttransfer_4K.append(np.mean(time_starttransfer_4K))
        mean_time_total_4K.append(np.mean(time_total_4K))
        
        # std
        std_time_namelookup_none.append(np.std(mean_time_namelookup_none))
        std_time_connect_none.append(np.std(mean_time_connect_none))
        std_time_pretransfer_none.append(np.std(mean_time_pretransfer_none))
        std_time_starttransfer_none.append(np.std(mean_time_starttransfer_none))
        std_time_total_none.append(np.std(mean_time_total_none))
        
        std_time_namelookup_360P.append(np.std(mean_time_namelookup_360P))
        std_time_connect_360P.append(np.std(mean_time_connect_360P)) 
        std_time_pretransfer_360P.append(np.std(mean_time_pretransfer_360P))
        std_time_starttransfer_360P.append(np.std(mean_time_starttransfer_360P)) 
        std_time_total_360P.append(np.std(mean_time_total_360P))
        
        std_time_namelookup_480P.append(np.std(mean_time_namelookup_480P)) 
        std_time_connect_480P.append(np.std(mean_time_connect_480P)) 
        std_time_pretransfer_480P.append(np.std(mean_time_pretransfer_480P))
        std_time_starttransfer_480P.append(np.std(mean_time_starttransfer_480P))
        std_time_total_480P.append(np.std(mean_time_total_480P))
        
        std_time_namelookup_HD.append(np.std(mean_time_namelookup_HD)) 
        std_time_connect_HD.append(np.std(mean_time_connect_HD))
        std_time_pretransfer_HD.append(np.std(mean_time_pretransfer_HD))
        std_time_starttransfer_HD.append(np.std(mean_time_starttransfer_HD)) 
        std_time_total_HD.append(np.std(mean_time_total_HD))
        
        std_time_namelookup_2K.append(np.std(mean_time_namelookup_2K)) 
        std_time_connect_2K.append(np.std(mean_time_connect_2K)) 
        std_time_pretransfer_2K.append(np.std(mean_time_pretransfer_2K))
        std_time_starttransfer_2K.append(np.std(mean_time_starttransfer_2K)) 
        std_time_total_2K.append(np.std(mean_time_total_2K))
        
        std_time_namelookup_4K.append(np.std(mean_time_namelookup_4K)) 
        std_time_connect_4K.append(np.std(mean_time_connect_4K))
        std_time_pretransfer_4K.append(np.std(mean_time_pretransfer_4K))
        std_time_starttransfer_4K.append(np.std(mean_time_starttransfer_4K)) 
        std_time_total_4K.append(np.std(mean_time_total_4K))
        
        
```
