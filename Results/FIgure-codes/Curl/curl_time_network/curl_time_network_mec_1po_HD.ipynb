```python
import matplotlib.pyplot as plot
import csv
import numpy as np
```

```python

# wifi
file_name = "../data/29_7_2023/wifi/mec/data/curl/mec/x86_pod_1_rep_1_27_7_2023_13h44.csv"

time_namelookup_wifi = []
time_connect_wifi = []
time_pretransfer_wifi = []
time_starttransfer_wifi = []
time_total_wifi = []


with open(file_name, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        if row[8] == "HD":
            time_namelookup_wifi.append(float(row[0]))
            time_connect_wifi.append(float(row[1]))
            time_pretransfer_wifi.append(float(row[3]))
            time_starttransfer_wifi.append(float(row[5]))
            time_total_wifi.append(float(row[6]))

    # Mean
    mean_time_namelookup_wifi = np.mean(time_namelookup_wifi)
    mean_time_connect_wifi = np.mean(time_connect_wifi)
    mean_time_pretransfer_wifi = np.mean(time_pretransfer_wifi)
    mean_time_starttransfer_wifi = np.mean(time_starttransfer_wifi)
    mean_time_total_wifi = np.mean(time_total_wifi)
    
    # std
    std_time_namelookup_wifi = np.std(time_namelookup_wifi)
    std_time_connect_wifi = np.std(time_connect_wifi)
    std_time_pretransfer_wifi = np.std(time_pretransfer_wifi)
    std_time_starttransfer_wifi = np.std(time_starttransfer_wifi)
    std_time_total_wifi = np.std(time_total_wifi)
           
        
```

```python
# 3G
file_name = "../data/29_7_2023/3G/mec/data/curl/mec/x86_pod_1_rep_1_28_7_2023_6h9.csv"

time_namelookup_3G = []
time_connect_3G = []
time_pretransfer_3G = []
time_starttransfer_3G = []
time_total_3G = []

with open(file_name, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        if row[8] == "HD":
            time_namelookup_3G.append(float(row[0]))
            time_connect_3G.append(float(row[1]))
            time_pretransfer_3G.append(float(row[3]))
            time_starttransfer_3G.append(float(row[5]))
            time_total_3G.append(float(row[6]))

    # Mean
    mean_time_namelookup_3G = np.mean(time_namelookup_3G)
    mean_time_connect_3G = np.mean(time_connect_3G)
    mean_time_pretransfer_3G = np.mean(time_pretransfer_3G)
    mean_time_starttransfer_3G = np.mean(time_starttransfer_3G)
    mean_time_total_3G = np.mean(time_total_3G)
    
    # std
    std_time_namelookup_3G = np.std(time_namelookup_3G)
    std_time_connect_3G = np.std(time_connect_3G)
    std_time_pretransfer_3G = np.std(time_pretransfer_3G)
    std_time_starttransfer_3G = np.std(time_starttransfer_3G)
    std_time_total_3G = np.std(time_total_3G)
```

```python

data_3G= [mean_time_namelookup_3G, mean_time_connect_3G, mean_time_pretransfer_3G, mean_time_starttransfer_3G, mean_time_total_3G]
data_4G = [mean_time_namelookup_4G, mean_time_connect_4G, mean_time_pretransfer_4G, mean_time_starttransfer_4G, mean_time_total_4G]
data_wifi = [mean_time_namelookup_wifi, mean_time_connect_wifi, mean_time_pretransfer_wifi, mean_time_starttransfer_wifi, mean_time_total_wifi]

error_data_3G= [std_time_namelookup_3G, std_time_connect_3G, std_time_pretransfer_3G, std_time_starttransfer_3G, std_time_total_3G]
error_data_4G = [std_time_namelookup_4G, std_time_connect_4G, std_time_pretransfer_4G, std_time_starttransfer_4G, std_time_total_4G]
error_data_wifi = [std_time_namelookup_wifi, std_time_connect_wifi, std_time_pretransfer_wifi, std_time_starttransfer_wifi, std_time_total_wifi]


time = ["time namelookup", "time connect", "time pretransfer", "time start tranfer", "time total"]

fig1 = plot.subplots(figsize =(15, 8))

barWidth = 0.25
# Set position of bar on X axis
br1 = np.arange(len(time))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plot.bar(br1, data_3G, color = 'darkblue', width = barWidth, label = '3G', yerr=error_data_3G, align='center', alpha=0.9, ecolor='red', capsize=12)
plot.bar(br2, data_4G, color = 'blue', width=barWidth, label = '4G', yerr=error_data_4G, align='center', alpha=0.9, ecolor='red', capsize=12)
plot.bar(br3, data_wifi, color = 'c', width=barWidth, label = 'wifi', yerr=error_data_wifi, align='center', alpha=0.9, ecolor='red', capsize=12)

# ax1.set_title('Average power usage of states')
# br1_round = [round(x, 2) for x in data_time_name_lookup]
# bars= plot.bar(quality, br1_round, color="darkblue")
# plot.bar_label(bars, label_type='center',color="w", fontsize = 14)

# Adding Xticks
plot.xlabel('Time components', fontsize = 15)
plot.ylabel('Time (s)', fontsize = 15)
plot.xticks([r + barWidth for r in range(len(time))], time)
plot.title(label="MEC - 1 pod - HD")
plot.ylim(0,3.5)
plot.legend()
# plot.savefig('../pdf/power-states.pdf', bbox_inches='tight')
plot.show()
```
