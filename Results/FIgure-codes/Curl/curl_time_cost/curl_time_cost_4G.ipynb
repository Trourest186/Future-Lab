```python
import matplotlib.pyplot as plot
import csv
import numpy as np
import random
```

```python
# 360P [1pod edge, 1pod mec, 2pods mec, 3 pods mec, 4pods mec, 5 pods mec]
mean_time_total_360P = []
# 480P [1pod edge, 1pod mec, 2pods mec, 3 pods mec, 4pods mec, 5 pods mec]
mean_time_total_480P = []
# HD [1pod edge, 1pod mec, 2pods mec, 3 pods mec, 4pods mec, 5 pods mec]
mean_time_total_HD = []
# 2K [1pod edge, 1pod mec, 2pods mec, 3 pods mec, 4pods mec, 5 pods mec]
mean_time_total_2K = []
```

```python
# 360P
std_time_total_360P = []
# 480P
std_time_total_480P = []
# HD
std_time_total_HD = []
# 2K
std_time_total_2K = []
```

```python
image_size_360p = 60 #KB
image_size_480p = 92 #KB
image_size_HD = 212 #KB
image_size_2K = 436 #KB
```

```python
# get edge data
file_name = "../data/29_7_2023/4G/jetson/data/curl/jetson/arm_pod_1_rep_1_29_7_2023_11h28.csv"
# 360P
time_total_360P = []
# 480P
time_total_480P = []
# HD
time_total_HD = []
# 2K
time_total_2K = []

with open(file_name, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        if row[8] == "360P":
            time_total_360P.append(float(row[6]))
        elif row[8] == "480P":
            time_total_480P.append(float(row[6]))
        elif row[8] == "HD":
            time_total_HD.append(float(row[6]))
        elif row[8] == "2K":
            time_total_2K.append(float(row[6]))

    # Mean
    mean_time_total_360P.append(np.mean(time_total_360P)/image_size_360p)

    mean_time_total_480P.append(np.mean(time_total_480P)/image_size_480p)

    mean_time_total_HD.append(np.mean(time_total_HD)/image_size_HD)

    mean_time_total_2K.append(np.mean(time_total_2K)/image_size_2K)
    
    # std

    std_time_total_360P.append(np.std(time_total_360P)/image_size_360p)

    std_time_total_480P.append(np.std(time_total_480P)/image_size_480p)
    
    std_time_total_HD.append(np.std(time_total_HD)/image_size_HD)
    
    std_time_total_2K.append(np.std(time_total_2K)/image_size_2K)
    
```

```python

for i in range(1,6):
    file_name = "../data/29_7_2023/4G/mec/data/curl/mec/x86_pod_"+str(i)+"_rep_1_27_7_2023_17h29.csv"
    # 360P
    time_total_360P = []
    # 480P
    time_total_480P = []
    # HD
    time_total_HD = []
    # 2K
    time_total_2K = []

    with open(file_name, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in spamreader:
            if row[8] == "360P":
                time_total_360P.append(float(row[6]))
            elif row[8] == "480P":
                time_total_480P.append(float(row[6]))
            elif row[8] == "HD":
                time_total_HD.append(float(row[6]))
            elif row[8] == "2K":
                time_total_2K.append(float(row[6]))

        # Mean
        mean_time_total_360P.append(np.mean(time_total_360P)/image_size_360p)

        mean_time_total_480P.append(np.mean(time_total_480P)/image_size_480p)

        mean_time_total_HD.append(np.mean(time_total_HD)/image_size_HD)

        mean_time_total_2K.append(np.mean(time_total_2K)/image_size_2K)
        
        # std

        std_time_total_360P.append(np.std(time_total_360P)/image_size_360p)

        std_time_total_480P.append(np.std(time_total_480P)/image_size_480p)
        
        std_time_total_HD.append(np.std(time_total_HD)/image_size_HD)
        
        std_time_total_2K.append(np.std(time_total_2K)/image_size_2K)
        
        
        
```

```python
data_1pod_edge = [mean_time_total_360P[0], mean_time_total_480P[0], mean_time_total_HD[0], mean_time_total_2K[0]]
data_1pod_mec = [mean_time_total_360P[1], mean_time_total_480P[1], mean_time_total_HD[1], mean_time_total_2K[1]]
data_2pod_mec = [mean_time_total_360P[2], mean_time_total_480P[2], mean_time_total_HD[2], mean_time_total_2K[2]]
data_3pod_mec = [mean_time_total_360P[3], mean_time_total_480P[3], mean_time_total_HD[3], mean_time_total_2K[3]]
data_4pod_mec = [mean_time_total_360P[4], mean_time_total_480P[4], mean_time_total_HD[4], mean_time_total_2K[4]]
data_5pod_mec = [mean_time_total_360P[5], mean_time_total_480P[5], mean_time_total_HD[5], mean_time_total_2K[5]]


error_data_1pod_edge = [std_time_total_360P[0], std_time_total_480P[0], std_time_total_HD[0], std_time_total_2K[0]]
error_data_1pod_mec = [std_time_total_360P[1], std_time_total_480P[1], std_time_total_HD[1], std_time_total_2K[1]]
error_data_2pod_mec = [std_time_total_360P[2], std_time_total_480P[2], std_time_total_HD[2], std_time_total_2K[2]]
error_data_3pod_mec = [std_time_total_360P[3], std_time_total_480P[3], std_time_total_HD[3], std_time_total_2K[3]]
error_data_4pod_mec = [std_time_total_360P[4], std_time_total_480P[4], std_time_total_HD[4], std_time_total_2K[4]]
error_data_5pod_mec = [std_time_total_360P[5], std_time_total_480P[5], std_time_total_HD[5], std_time_total_2K[5]]

quality = ["360P", "480P", "HD", "2K"]

fig1 = plot.subplots(figsize =(30, 8))

barWidth = 0.15
# Set position of bar on X axis
br1 = np.arange(len(quality))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
br6 = [x + barWidth for x in br5]

plot.bar(br1, data_1pod_edge, color = 'yellow', width = barWidth, label = '1 pod edge', yerr=error_data_1pod_edge, align='center', alpha=0.9, ecolor='red', capsize=12)
plot.bar(br2, data_1pod_mec, color = 'orange', width=barWidth, label = '1 pod mec', yerr=error_data_1pod_mec, align='center', alpha=0.9, ecolor='red', capsize=12)
plot.bar(br3, data_2pod_mec, color = 'g', width=barWidth, label = '2 pods mec', yerr=error_data_2pod_mec, align='center', alpha=0.9, ecolor='red', capsize=12)
plot.bar(br4, data_3pod_mec, color = 'c', width=barWidth, label = '3 pods mec', yerr=error_data_3pod_mec, align='center', alpha=0.9, ecolor='red', capsize=12)
plot.bar(br5, data_4pod_mec, color = 'b', width=barWidth, label = '4 pods mec', yerr=error_data_4pod_mec, align='center', alpha=0.9, ecolor='red', capsize=12)
plot.bar(br6, data_5pod_mec, color = 'darkblue', width=barWidth, label = '5 pods mec', yerr=error_data_5pod_mec, align='center', alpha=0.9, ecolor='red', capsize=12)

# ax1.set_title('Average power usage of states')
# br1_round = [round(x, 2) for x in data_time_name_lookup]
# bars= plot.bar(quality, br1_round, color="darkblue")
# plot.bar_label(bars, label_type='center',color="w", fontsize = 14)

# Adding Xticks
# plot.xlabel('Branch', fontweight ='bold', fontsize = 15)
plot.ylabel('Cost (s/KB)', fontsize = 15)
plot.xticks([r + barWidth for r in range(len(quality))], quality)
plot.title("Cost - 4G", fontsize = 30)
plot.legend()
# plot.savefig('../pdf/power-states.pdf', bbox_inches='tight')
plot.show()
```
