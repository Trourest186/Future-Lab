```python
# import
import matplotlib.pyplot as plot
import csv
import numpy as np
import random
import glob
import scipy.stats
# Calculate z-value for 90% confidence interval
confidence_level = 0.90
z_value = scipy.stats.norm.ppf((1 + confidence_level) / 2)  # For 90% confidence interval
```

```python
# Jetson resources
average_ram_null_state_edge = []
average_cpu_null_state_edge = []
average_ram_cold_state_edge = []
average_cpu_cold_state_edge = []
average_ram_warm_disk_state_edge = []
average_cpu_warm_disk_state_edge = []
average_ram_warm_cpu_state_edge = []
average_cpu_warm_cpu_state_edge = []
average_ram_active_state_edge = []
average_cpu_active_state_edge = []


for i in range(1,11):
    file_name = "../../../Measurements/Lifecycle/wifi/resource/jetson/arm_pod_1_rep_"+str(i)+"_18_7_2023_8h55.csv"
    #STATES
    ram_null_state_edge = []
    cpu_null_state_edge = []
    ram_warm_disk_state_edge = []
    cpu_warm_disk_state_edge = []
    ram_warm_cpu_state_edge = []
    cpu_warm_cpu_state_edge = []
    ram_active_state_edge = []
    cpu_active_state_edge = []
    ram_cold_state_edge = []
    cpu_cold_state_edge = []

    with open(file_name, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in spamreader:
            if row[9] == "null_state":
                cpu_null_state_edge.append(float(row[5]))
                ram_null_state_edge.append(float(row[7]))
            elif row[9] == "warm_disk_state":
                cpu_warm_disk_state_edge.append(float(row[5]))
                ram_warm_disk_state_edge.append(float(row[7]))
            elif row[9] == "warm_cpu_state":
                cpu_warm_cpu_state_edge.append(float(row[5]))
                ram_warm_cpu_state_edge.append(float(row[7]))
            elif row[9] == "active_state":
                cpu_active_state_edge.append(float(row[5]))
                ram_active_state_edge.append(float(row[7]))
            elif row[9] == "cold_state":
                cpu_cold_state_edge.append(float(row[5]))
                ram_cold_state_edge.append(float(row[7]))
    
        #STATES
        average_cpu_null_state_edge.append(np.mean(cpu_null_state_edge))
        average_ram_null_state_edge.append(np.mean(ram_null_state_edge))

        average_cpu_warm_disk_state_edge.append(np.mean(cpu_warm_disk_state_edge))
        average_ram_warm_disk_state_edge.append(np.mean(ram_warm_disk_state_edge))

        average_cpu_warm_cpu_state_edge.append(np.mean(cpu_warm_cpu_state_edge))
        average_ram_warm_cpu_state_edge.append(np.mean(ram_warm_cpu_state_edge))

        average_cpu_active_state_edge.append(np.mean(cpu_active_state_edge))
        average_ram_active_state_edge.append(np.mean(ram_active_state_edge))

        average_cpu_cold_state_edge.append(np.mean(cpu_cold_state_edge))
        average_ram_cold_state_edge.append(np.mean(ram_cold_state_edge))

#STATES
# Calculate the average
ram_null_state_edge_mean = np.mean(average_ram_null_state_edge)
cpu_null_state_edge_mean = np.mean(average_cpu_null_state_edge)

ram_warm_disk_state_edge_mean = np.mean(average_ram_warm_disk_state_edge)
cpu_warm_disk_state_edge_mean = np.mean(average_cpu_warm_disk_state_edge) 

ram_warm_cpu_state_edge_mean = np.mean(average_ram_warm_cpu_state_edge) 
cpu_warm_cpu_state_edge_mean = np.mean(average_cpu_warm_cpu_state_edge)

ram_active_state_edge_mean = np.mean(average_ram_active_state_edge)
cpu_active_state_edge_mean = np.mean(average_cpu_active_state_edge)

ram_cold_state_edge_mean = np.mean(average_ram_cold_state_edge)
cpu_cold_state_edge_mean = np.mean(average_cpu_cold_state_edge)

# Calculate the standard deviation
ram_null_state_edge_std = np.std(average_ram_null_state_edge) 
cpu_null_state_edge_std = np.std(average_cpu_null_state_edge) 

ram_warm_disk_state_edge_std = np.std(average_ram_warm_disk_state_edge)/np.sqrt(len(average_ram_warm_disk_state_edge))
cpu_warm_disk_state_edge_std = np.std(average_cpu_warm_disk_state_edge)/np.sqrt(len(average_cpu_warm_disk_state_edge))

ram_warm_cpu_state_edge_std = np.std(average_ram_warm_cpu_state_edge)/np.sqrt(len(average_ram_warm_cpu_state_edge))
cpu_warm_cpu_state_edge_std = np.std(average_cpu_warm_cpu_state_edge)/np.sqrt(len(average_cpu_warm_cpu_state_edge))

ram_active_state_edge_std = np.std(average_ram_active_state_edge)/np.sqrt(len(average_ram_active_state_edge))
cpu_active_state_edge_std = np.std(average_cpu_active_state_edge)/np.sqrt(len(average_cpu_active_state_edge))

ram_cold_state_edge_std = np.std(average_ram_cold_state_edge)/np.sqrt(len(average_ram_cold_state_edge))
cpu_cold_state_edge_std = np.std(average_cpu_cold_state_edge)/np.sqrt(len(average_cpu_cold_state_edge))
```

```python
# MEC resources

# Jetson resources
average_ram_null_state_mec = []
average_cpu_null_state_mec = []
average_ram_cold_state_mec = []
average_cpu_cold_state_mec = []
average_ram_warm_disk_state_mec = []
average_cpu_warm_disk_state_mec = []
average_ram_warm_cpu_state_mec = []
average_cpu_warm_cpu_state_mec = []
average_ram_active_state_mec = []
average_cpu_active_state_mec = []


for i in range(1,11):
    file_name = "../../../Measurements/Lifecycle/wifi/resource/mec/x86_pod_5_rep_"+str(i)+"_12_7_2023_12h48.csv"
    #STATES
    ram_null_state_mec = []
    cpu_null_state_mec = []
    ram_warm_disk_state_mec = []
    cpu_warm_disk_state_mec = []
    ram_warm_cpu_state_mec = []
    cpu_warm_cpu_state_mec = []
    ram_active_state_mec = []
    cpu_active_state_mec = []
    ram_cold_state_mec = []
    cpu_cold_state_mec = []

    with open(file_name, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in spamreader:
            if row[9] == "null_state":
                cpu_null_state_mec.append(float(row[5]))
                ram_null_state_mec.append(float(row[7]))
            elif row[9] == "warm_disk_state":
                cpu_warm_disk_state_mec.append(float(row[5]))
                ram_warm_disk_state_mec.append(float(row[7]))
            elif row[9] == "warm_cpu_state":
                cpu_warm_cpu_state_mec.append(float(row[5]))
                ram_warm_cpu_state_mec.append(float(row[7]))
            elif row[9] == "active_state":
                cpu_active_state_mec.append(float(row[5]))
                ram_active_state_mec.append(float(row[7]))
            elif row[9] == "cold_state":
                cpu_cold_state_mec.append(float(row[5]))
                ram_cold_state_mec.append(float(row[7]))
    
        #STATES
        average_cpu_null_state_mec.append(np.mean(cpu_null_state_mec))
        average_ram_null_state_mec.append(np.mean(ram_null_state_mec))

        average_cpu_warm_disk_state_mec.append(np.mean(cpu_warm_disk_state_mec))
        average_ram_warm_disk_state_mec.append(np.mean(ram_warm_disk_state_mec))

        average_cpu_warm_cpu_state_mec.append(np.mean(cpu_warm_cpu_state_mec))
        average_ram_warm_cpu_state_mec.append(np.mean(ram_warm_cpu_state_mec))

        average_cpu_active_state_mec.append(np.mean(cpu_active_state_mec))
        average_ram_active_state_mec.append(np.mean(ram_active_state_mec))

        average_cpu_cold_state_mec.append(np.mean(cpu_cold_state_mec))
        average_ram_cold_state_mec.append(np.mean(ram_cold_state_mec))

#STATES
# Calculate the average
ram_null_state_mec_mean = np.mean(average_ram_null_state_mec)
cpu_null_state_mec_mean = np.mean(average_cpu_null_state_mec)

ram_warm_disk_state_mec_mean = np.mean(average_ram_warm_disk_state_mec)
cpu_warm_disk_state_mec_mean = np.mean(average_cpu_warm_disk_state_mec) 

ram_warm_cpu_state_mec_mean = np.mean(average_ram_warm_cpu_state_mec) 
cpu_warm_cpu_state_mec_mean = np.mean(average_cpu_warm_cpu_state_mec)

ram_active_state_mec_mean = np.mean(average_ram_active_state_mec)
cpu_active_state_mec_mean = np.mean(average_cpu_active_state_mec)

ram_cold_state_mec_mean = np.mean(average_ram_cold_state_mec)
cpu_cold_state_mec_mean = np.mean(average_cpu_cold_state_mec)

# Calculate the standard deviation
ram_null_state_mec_std = np.std(average_ram_null_state_mec) 
cpu_null_state_mec_std = np.std(average_cpu_null_state_mec) 

ram_warm_disk_state_mec_std = np.std(average_ram_warm_disk_state_mec)/np.sqrt(len(average_ram_warm_disk_state_mec))
cpu_warm_disk_state_mec_std = np.std(average_cpu_warm_disk_state_mec)/np.sqrt(len(average_cpu_warm_disk_state_mec))

ram_warm_cpu_state_mec_std = np.std(average_ram_warm_cpu_state_mec)/np.sqrt(len(average_ram_warm_cpu_state_mec))
cpu_warm_cpu_state_mec_std = np.std(average_cpu_warm_cpu_state_mec)/np.sqrt(len(average_cpu_warm_cpu_state_mec))

ram_active_state_mec_std = np.std(average_ram_active_state_mec)/np.sqrt(len(average_ram_active_state_mec))
cpu_active_state_mec_std = np.std(average_cpu_active_state_mec)/np.sqrt(len(average_cpu_active_state_mec))

ram_cold_state_mec_std = np.std(average_ram_cold_state_mec)/np.sqrt(len(average_ram_cold_state_mec))
cpu_cold_state_mec_std = np.std(average_cpu_cold_state_mec)/np.sqrt(len(average_cpu_cold_state_mec))
```

```python
# Plotting functions CPU

tick_fontsize = 20
bar_fontsize = 18
cap_size = 14

def custom_edge(ax = None):
    if ax is None:
        fig, ax = plot.subplots()
    data = [cpu_null_state_edge_mean, cpu_cold_state_edge_mean,  cpu_warm_disk_state_edge_mean, cpu_warm_cpu_state_edge_mean,
        cpu_active_state_edge_mean]
    error = [cpu_null_state_edge_std-5, cpu_cold_state_edge_std,  cpu_warm_disk_state_edge_std, cpu_warm_cpu_state_edge_std,
        cpu_active_state_edge_std]
    errorx = [x*z_value for x in error]

# state = ["$S_n$", "$S_c$", "$S_{wd}$", "$S_{wc}$", "$S_{wm}$",  "$S_a$"]
    state = ["Null", "Cold", "Warm Disk", "Warm CPU", "Active"]

    x_pos = np.arange(len(state))
    ax.bar(x_pos, data, yerr=errorx, align='center', alpha=0.9, ecolor='red', capsize=cap_size)
    # ax1.set_title('RAM of states')
    data_round = [round(x, 2) for x in data]
    bars= ax.bar(state, data_round, color="darkblue")
    ax.bar_label(bars, label_type='center',color="w", fontsize = bar_fontsize)
    ax.set_ylabel("CPU Edge(%)", fontsize = tick_fontsize)
    # ax1.set_xlabel("State ($S$)",  fontsize = 15)
    ax.set_xticks(x_pos)
    ax.tick_params(axis='both', which='major', labelsize=tick_fontsize)

    ax.set_xticklabels(state, rotation = tick_fontsize)
    ax.set_yticks(np.linspace(0,100,6))
    # ax1.set_xticklabels(ax1.get_xticks(), )

# ax1.tick_params(axis='both', which='major', labelsize=12)
#     x1 = np.linspace(0.0, 5.0)
#     y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
#     ax.plot(x1, y1, 'ko-')
#     ax.set_xlabel('time (s)')
#     ax.set_ylabel('Damped oscillation')

def custom_mec(ax = None):
    if ax is None:
        fig, ax = plot.subplots()
    data = [cpu_null_state_mec_mean, cpu_cold_state_mec_mean,  cpu_warm_disk_state_mec_mean, cpu_warm_cpu_state_mec_mean,
        cpu_active_state_mec_mean]
    error = [cpu_null_state_mec_std, cpu_cold_state_mec_std,  cpu_warm_disk_state_mec_std, cpu_warm_cpu_state_mec_std,
       cpu_active_state_mec_std]
    errorx = [x*z_value for x in error]

# state = ["$S_n$", "$S_c$", "$S_{wd}$", "$S_{wc}$", "$S_{wm}$",  "$S_a$"]
    state = ["Null", "Cold", "Warm Disk", "Warm CPU", "Active"]

    x_pos = np.arange(len(state))
    ax.bar(x_pos, data, yerr=errorx, align='center', alpha=0.9, ecolor='red', capsize=cap_size)
    # ax1.set_title('RAM of states')
    data_round = [round(x, 2) for x in data]
    bars= ax.bar(state, data_round, color="darkblue")
    ax.tick_params(axis='both', which='major', labelsize=tick_fontsize)
    ax.bar_label(bars, label_type='center',color="w", fontsize = bar_fontsize)
    ax.set_ylabel("CPU MEC(%)",  fontsize = tick_fontsize)
    # ax1.set_xlabel("State ($S$)",  fontsize = 15)
    ax.set_xticks(x_pos)
    ax.set_yticks(np.linspace(0,100,6))
    ax.set_xticklabels(state, rotation = tick_fontsize)
```

```python
# Start plotting

fig = plot.figure(figsize = (15,5.5))
# ax1 = fig.add_subplot(1,2,1)
# custom_edge(ax1)
ax2 = fig.add_subplot(1,2,2)
custom_mec(ax2)
# plot.savefig('./new/RAM-states-mec.pdf', bbox_inches='tight')
plot.savefig('./new/CPU-states-mec.pdf', bbox_inches='tight')
```
