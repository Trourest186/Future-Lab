```python
import matplotlib.pyplot as plot
import csv
import numpy as np
import random
import glob
```

```python
#Warm disk to warm cpu
# [1 pod, 2p, 3p, 4p, 5p]
# mean
mean_time_3G = [] 
mean_time_4G = [] 
mean_time_wifi = [] 
# std
std_time_3G = [] 
std_time_4G = [] 
std_time_wifi = [] 
```

```python
#3G
average_time = [] 

for j in range(1, 11):
    file_name_list = glob.glob("../../Measurements/Lifecycle/3G/timestamp/jetson/time_arm_pod_1_rep_"+str(j)+"_*")
    # print(len(file_name_list))
    # if len(file_name_list) == 2:
    #     print(file_name_list)
    #PROCESS
    time_warm_disk_to_warm_cpu_process = []
    for file_name in file_name_list:
        time_warm_disk_to_warm_cpu_process = []
        with open(file_name, newline='') as csvfile:
            spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
            for row in spamreader:
                if row[2] == "warm_disk_to_warm_CPU_process":
                    time_warm_disk_to_warm_cpu_process.append(float(row[1]))
            average_time.append(time_warm_disk_to_warm_cpu_process[-1]-time_warm_disk_to_warm_cpu_process[0])
mean_time_3G.append(np.mean(average_time))
std_time_3G.append(np.std(average_time))


for i in range(1,6):
    average_time = [] 
    for j in range(1, 11):
        file_name_list = glob.glob("../../Measurements/Lifecycle/3G/timestamp/mec/time_x86_pod_"+str(i)+"_rep_"+str(j)+"_*")
        # print(len(file_name_list))
        # if len(file_name_list) == 2:
        #     print(file_name_list)
        #PROCESS
        for file_name in file_name_list:
            time_warm_disk_to_warm_cpu_process = []
            with open(file_name, newline='') as csvfile:
                spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
                for row in spamreader:
                    if row[2] == "warm_disk_to_warm_CPU_process":
                        time_warm_disk_to_warm_cpu_process.append(float(row[1]))
                average_time.append(time_warm_disk_to_warm_cpu_process[-1]-time_warm_disk_to_warm_cpu_process[0])
    mean_time_3G.append(np.mean(average_time))
    std_time_3G.append(np.std(average_time))
```
